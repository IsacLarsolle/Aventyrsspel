3.1 Äventyrsspelet
Ni ska skapa ett textbaserat äventyrsspel där en äventyrare navigerar genom rum, samlar på sig föremål och slåss mot monster. 
Spelet kommer att köras i en loop som i varje iteration frågar vad äventyraren vill göra. 
Användaren ska vid varje iteration kunna
    Visa äventyrarens egenskaper
        -HP (hälsopoäng)
        -STR (styrka)
        -LVL (nivå)
    -Visa innehållet i äventyrarens inventory (ryggsäck)
    -Välja en av tre dörrar

Bakom varje dörr kan det antingen gömma sig ett monster, en kista eller en fälla. 
Vad som finns bakom varje dörr ska vara slumpmässigt.

3.1.1 Monster
Om äventyraren stöter på ett monster kommer en strid att utspela sig. 
Ett monster har en styrka (STR) som slumpas fram. 
Äventyrarens STR i striden beräknas genom att summera äventyrarens STR och eventuella bonusar från föremål i inventoryt.

Striden har följande regler:
    -Om monstret har mer STR än äventyraren vinner monstret och äventyraren 
    förlorar 1 HP. (Om äventyrarens HP går ner till 0 är spelet över.)
    -Om äventyraren har mer STR är monstret vinner äventyraren. Monstret flyr och äventyraren går upp 1 LVL. (Då äventyrarens LVL når 10 är spelet vunnet.)
    -Om äventyraren och monstret har samma STR händer ingenting.

Hur striden går ska skrivas ut så att användaren kan se. 
Efter att striden är avslutad får användaren än en gång välja vad den ska göra, d.v.s. kolla sitt inventory, sina egenskaper eller välja en av tre nya dörrar.

3.1.2 Kistor
En kista innehåller ett föremål som äventyraren kan lägga till sitt inventory. 
Varje föremål har en slumpad STR-bonus som används i strid. 
Om en äventyrare hittar en kista ska det skrivas ut att en kista hittats och vad den innehåller. 
Föremålet läggs i äventyrarens inventory och ger äventyraren STR-bonus i framtida strider. 
Kom ihåg att räkna med eventuella föremål i äventyrarens inventory när striders utfall ska beräknas. 
Efter att föremålet lagts till inventoryt får användaren än en gång välja vad den ska göra, d.v.s. kolla sitt inventory, sina egenskaper eller välja av tre nya dörrar.

3.1.3 Fällor
Stöter äventyraren på en fälla så förlorar den HP. Skadan en fälla gör slumpas fram av spelet.
Det ska skrivas ut att äventyraren stöter på en fälla samt hur mycket skada äventyraren tar.
Om äventyrarens HP går ner till 0 är spelet över.

Efter att fällan drabbat äventyraren får användaren än en gång välja vad den ska göra, d.v.s. kolla sitt inventory, sina egenskaper eller välja av tre nya dörrar.

3.1.4 Krav för betyget E
    -Med ert program ska användaren kunna:
        -Välja att gå in i ett av tre rum (framåt, höger, vänster).
        -Välja att se information om sig själv (INV, HP, STR).
        -När äventyraren går in i ett rum med ett monster ska en strid utkämpas. Se regler för strid ovan.
            -Om äventyrarens HP når noll är spelet över.
        -När äventyraren går in i ett rum med en fälla ska äventyraren förlora HP. 
            -Om äventyrarens HP når noll är spelet över.
        -När äventyraren går in i ett rum med en kista ska föremålet i kistan läggas till äventyrarens inventory.

3.1.5 Ytterligare krav för betyget C
    -Felhantering:
        -Programmet ska hantera om användaren skriver in andra menyval än de beskrivna, 
        som “H”, “12345” eller annan inmatning.
    -Spelare ska kunna klättra i rang (level), ifall de besegrar monster eller plockar upp skatter, och klara spelet ifall de når upp till rang 10. 
    Om spelaren väljer menyalternativet för att skriva ut information om spelarens karaktär så ska rang skrivas ut.

3.1.6 Ytterligare krav för betyget A
    -Du ska använda objektorientering.
        -Du ska skapa klassen Player som har attributen: strength, hp, inventory och level.
        -Inventory ska vara en lista av klassen Item (som du skapar). Item ska ha attributen 
        name och strength_bonus.
        -När inventoryt skrivs ut ska både name och strength_bonus skrivas ut för varje föremål.
        -I kistorna ska det finnas ett objekt av klassen Item med name och slumpad strength_bonus.
    -Äventyrarens inventory rymmer endast 5 föremål. Om inventoryt är fullt och äventyraren hittar en kista ska spelaren få välja om den vill byta ut ett föremål i inventoryt och i så fall ange vilket föremål som ska bytas ut.
